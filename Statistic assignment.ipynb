```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Laliga.csv',skiprows=[0])
```

```python
df
```

```python
df.replace(['-'],['0'],inplace=True)
```

```python
df
```

```python
df['Debut']=df['Debut'].astype(str).str[:4].astype(int)
df1 = df[df['Debut'].between(1930,1980)]
df1[['Team','Debut']]
```

```python
df['Points']=df['Points'].astype(int)
df.sort_values(by='Points',ascending = False).head(5)
```

```python
def Goal_diff_count():
    df['Goal_diff_count'] = df['GoalsFor'].astype(int)-df['GoalsAgainst'].astype(int)
    return df[['Team','Goal_diff_count']].sort_values(by ='Goal_diff_count', ascending = False)
Goal_diff_count()
```
